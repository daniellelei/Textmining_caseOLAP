```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("desktop/4_1caseolap.csv",index_col=0)
data
```

```python
data['IHD'] = pd.to_numeric(data.IHD)
data['CM'] = pd.to_numeric(data.CM)
data['ARR'] = pd.to_numeric(data.ARR)
data['VD'] = pd.to_numeric(data.VD)
data['CHD'] = pd.to_numeric(data.CHD)
data['CCD'] = pd.to_numeric(data.CCD)
data['VOO'] = pd.to_numeric(data.VOO)
data['OTH'] = pd.to_numeric(data.OTH)
pd.options.display.float_format = "{:,.2f}".format
new_data = data.query("IHD>0.1|CM>0.1|ARR>0.1|VD>0.1|CHD>0.1|CCD>0.1|VOO>0.1|OTH>0.1")
#new_data = data. query("IHD > 0.00533 & IHD < 0.00538")
new_data
```

```python
fig, ax = plt.subplots(figsize=(12,13))
sns.heatmap(new_data,cmap="YlGnBu", annot=True, cbar_kws = {'label':'color key','orientation':"vertical"})
ax.xaxis.tick_top()
xticks_labels = ['IHD','CM','ARR','VD','CHD','CCD','VOO','OTH']
plt.xticks(np.arange(8) + .5, labels=xticks_labels)
title = 'human-heart tissue specific-caseolap\n'.upper()
plt.title(title, loc = 'left')
plt.show()
```
