```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_mouseIHD = pd.read_csv('desktop/mousecaseolap.csv', index_col=0)

index_mIHD = data_mouseIHD.index
data_mouseIHD['ID']=index_mIHD
data_mouseIHD['ID']= data_mouseIHD['ID'].str.upper()

def search_ (df1, df2):
    result_found = 0
    df3 = pd.DataFrame()
    
    for i in range (len(df2.index)):
        mouse_ID = df2['ID'].values[i]        
        if df1['Cluster members'].str.contains(mouse_ID).any():
            df3 = df3.append(df2[i:i+1])
            result_found += 1
            
    display(df3) 
    #df3.to_csv('desktop/mouseIHD1.csv')
    return df3
```

```python
data_IHD = pd.read_excel('desktop/data1.xlsx', index_col=0)
similarity = search_ (data_IHD, data_mouseIHD)
similarity.to_csv('desktop/referencesimilarity.csv')
```
